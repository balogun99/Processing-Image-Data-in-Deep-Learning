# Resizing the Image using the Pillow Library

```python
# resizing the image
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')
image_resized = img.resize((200,200))
```

```python
image_resized.save('/content/dog.jpg')
```

```python
# displaying the image from a numpy array
image_resized = mpimg.imread('/content/dog.jpg')
image_resized_plot = plt.imshow(image_resized)
plt.show()
```

```python
print(image_resized.shape)
```

# Converting RGB images to Grayscale image using CV2

```python
import cv2
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

# cv2.imshow() will display the image. But this would not be allowed in the Google Colab.

```python
from google.colab.patches import cv2_imshow
```

```python
# display the image
cv2_imshow(grayscale_image)
```

```python
# saving the grayscale image
cv2.imwrite('/content/dog.jpg', grayscale_image)
```
